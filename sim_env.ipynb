```python
from Robot import Diif_Robot
import numpy as np
import matplotlib.pyplot as plt 

robot_initial_state = np.array([0,0,np.pi/2])

sample_robot = Diif_Robot(0.75,0.3/2,robot_initial_state)

tf = 20 
t = np.arange(0,tf,0.1)
dt = 0.1

all_x = []
all_y = []

for i in range(len(t)):
    # state = sample_robot.forward_kinematics(0.5,0.5,dt)
    u = np.array([0.5,0])
    state = sample_robot.motion_model(u,dt)
    all_x.append(state[1])
    all_y.append(state[0])
    # print(state[2])
    # plt.plot(sample_robot.x,sample_robot.y,'ro')
    # plt.pause(0.1)
plt.plot(all_x,all_y,'ro',markersize=1)
plt.xlim(-2,2)
plt.ylim(-2,2)
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
plt.grid()
plt.show() 
```
